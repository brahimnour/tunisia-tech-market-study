# 🔍 Étude de marché — Secteur tech tunisien
**Objectif :** Analyser les tendances du marché tech tunisien pour identifier
les secteurs porteurs et les profils les plus demandés en BI & Data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
print("Bibliothèques chargées ✓")
```

```python
url = "https://raw.githubusercontent.com/brahimnour/tunisia-tech-market-study/main/data/tunisia_tech_companies.csv"
df = pd.read_csv(url)

print("Données chargées ✓")
print("Dimensions :", df.shape)
df.head()
```

```python
print("Valeurs manquantes :")
print(df.isnull().sum())
print("\nNombre d'entreprises :", len(df))
print("Secteurs :", df['secteur'].unique())
```

```python
secteurs = df['secteur'].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=secteurs.values, y=secteurs.index, palette='Blues_r')
plt.title('Nombre d\'entreprises par secteur', fontsize=14, fontweight='bold')
plt.xlabel('Nombre d\'entreprises')
plt.ylabel('Secteur')
plt.tight_layout()
plt.savefig('secteurs.png', dpi=150)
plt.show()
```

```python
stages = df['stage_disponible'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(stages.values, labels=stages.index, autopct='%1.1f%%',
        colors=['#2E86AB', '#E84855'], startangle=90)
plt.title('Disponibilité des stages', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('stages.png', dpi=150)
plt.show()
```

```python
tech_list = []
for techs in df['technologies']:
    tech_list.extend(techs.split(';'))

tech_count = pd.Series(tech_list).value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=tech_count.values, y=tech_count.index, palette='viridis')
plt.title('Technologies les plus demandées', fontsize=14, fontweight='bold')
plt.xlabel('Nombre d\'entreprises')
plt.ylabel('Technologie')
plt.tight_layout()
plt.savefig('technologies.png', dpi=150)
plt.show()
```

```python
profils_list = []
for profils in df['profils_recherches']:
    profils_list.extend(profils.split(';'))

profils_count = pd.Series(profils_list).value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=profils_count.values, y=profils_count.index, palette='rocket')
plt.title('Profils les plus recherchés', fontsize=14, fontweight='bold')
plt.xlabel('Nombre d\'entreprises')
plt.ylabel('Profil')
plt.tight_layout()
plt.savefig('profils.png', dpi=150)
plt.show()
```

## 📌 Conclusions

- **Secteur dominant** : Consulting IT concentre le plus d'entreprises tech
- **Tunis** reste la ville principale pour les opportunités tech
- **SQL et Python** sont les technologies les plus demandées — cohérent avec mon profil
- **Plus de 70%** des entreprises proposent des stages
- **Data Analyst** est le profil le plus recherché sur le marché tunisien
